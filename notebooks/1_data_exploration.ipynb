```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
numerical_cols = df.select_dtypes(include='number').columns

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.tight_layout()
    plt.show()
```

MAXIMUM SCORE OF STUDENT IN ALL THREE SUBJECTS

```python
plt.figure(figsize=(18, 8))

plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.histplot(data=df, x='math score', color='red', kde=True, bins=20)

plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.histplot(data=df, x='reading score', color='green', kde=True, bins=20)

plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.histplot(data=df, x='writing score', color='blue', kde=True, bins=20)

plt.tight_layout()
plt.show()
```
